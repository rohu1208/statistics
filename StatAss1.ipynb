Q1: What are the Probability Mass Function (PMF) and Probability Density Function (PDF)? Explain with an example.

The Probability Mass Function (PMF) and Probability Density Function (PDF) are used to describe the probability distribution of a random variable.

PMF: The PMF is a function that assigns a probability to each possible value of a discrete random variable. It is denoted by P(x) and satisfies the following properties:

0 ≤ P(x) ≤ 1 for all x
∑P(x) = 1, where the sum is taken over all possible values of x
Example: Suppose we have a fair six-sided die. The PMF of the outcome X is:

P(X = 1) = 1/6, P(X = 2) = 1/6,..., P(X = 6) = 1/6

PDF: The PDF is a function that describes the probability distribution of a continuous random variable. It is denoted by f(x) and satisfies the following properties:

f(x) ≥ 0 for all x
∫f(x)dx = 1, where the integral is taken over all possible values of x
Example: Suppose we have a random variable X that follows a uniform distribution between 0 and 1. The PDF of X is:

f(x) = 1, for 0 ≤ x ≤ 1 f(x) = 0, otherwise

Q2: What is Cumulative Density Function (CDF)? Explain with an example. Why CDF is used?

The Cumulative Density Function (CDF) is a function that describes the cumulative probability of a random variable up to a certain point. It is denoted by F(x) and is defined as:

F(x) = P(X ≤ x)

The CDF is used to find the probability that a random variable takes on a value less than or equal to a certain value.

Example: Suppose we have a random variable X that follows a uniform distribution between 0 and 1. The CDF of X is:

F(x) = x, for 0 ≤ x ≤ 1 F(x) = 0, for x < 0 F(x) = 1, for x > 1

The CDF is used to find the probability that a random variable takes on a value within a certain range. For example, to find the probability that X is less than or equal to 0.5, we can use the CDF: P(X ≤ 0.5) = F(0.5) = 0.5.

Q9: What is Central Limit Theorem? State the significance of the Central Limit Theorem.

The Central Limit Theorem (CLT) is a fundamental concept in probability theory that states that, given certain conditions, the distribution of the mean of a large sample of independent and identically distributed random variables will be approximately normal, even if the underlying distribution is not normal.

The significance of the CLT lies in its far-reaching implications:

It provides a theoretical basis for many statistical methods and techniques, such as hypothesis testing and confidence intervals.
It allows us to make inferences about a population based on a sample of data.
It provides a way to approximate complex distributions using the normal distribution.
It has numerous applications in fields such as engineering, economics, and medicine.
Q10: State the assumptions of the Central Limit Theorem.

The Central Limit Theorem assumes that:

The random variables are independent and identically distributed (i.i.d.).
The random variables have a finite mean and variance.
The sample size is sufficiently large (typically n > 30).
These assumptions ensure that the CLT can be applied to a wide range of distributions, and that the resulting normal approximation is accurate.
